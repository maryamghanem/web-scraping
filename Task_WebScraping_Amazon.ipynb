```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
fathi=webdriver.Chrome()
```

```python
chromedriver_path = "C:\\Users\\alrhma\\Desktop\\WebScraping\\chromedriver.exe"
```

```python
url = " https://www.amazon.eg/s?k=samsung&rh=p_89%3Asamsung&language=en&ref=SQEG-WEB-SR301"
```

```python
def scrape_product_details(url):

    try:
        fathi.get(url)

        data = {
            "title": [],
            "image_url": [],
            "price": [],
            "rating": [],
            "review_count": []
        }

        # Extract product details using XPaths
        product_elements = fathi.find_elements('xpath', ".//div[@class='a-section a-spacing-base']")  

        for product in product_elements:

            try:
            
                title_element = product.find_element(By.XPATH, ".//h2/span[@class='a-size-base-plus a-color-base']")
                image_url_element = product.find_element(By.XPATH, ".//a/div/img[@class='s-image']")
                price_element = product.find_element(By.XPATH, ".//span[@class='a-price-whole']")
                rating_element = product.find_element(By.XPATH, ".//a/i/span[@class='a-icon-alt']")
                review_count_element = product.find_element(By.XPATH, ".//a/span[@class='a-size-base s-underline-text']") 
            

                data["title"].append(title_element.text)
                data["image_url"].append(image_url_element.get_attribute("src"))
                data["price"].append(price_element.text)
                data["rating"].append(rating_element.get_attribute('innerHTML').split(' ')[0]) 
                data["review_count"].append(review_count_element.text.strip()) 

            except Exception as e:
                print(f"Error extracting data for a product: {e}")

        # Creating a DataFrame
        df = pd.DataFrame(data)
        print(df.to_string())

        # Saving to a CSV file
        df.to_csv("amazon_data.csv", header=True, index=False)
        print(f"Scraped data for {len(df)} products")
        print("Data saved to amazon_data.csv")

    except Exception as e:
        print(f"Error scraping product details: {e}")
    finally:
        fathi.quit()

if __name__ == "__main__":
    scrape_product_details(url)
```

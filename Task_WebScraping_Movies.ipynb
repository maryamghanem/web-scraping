```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
fathi=webdriver.Chrome()

fathi.get('https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW')
```

```python
table =  fathi.find_elements('xpath',"//div[@class='a-section imdb-scroll-table-inner']")
for i in table:
     print(i.text)
```

```python
title_list = []
title =  fathi.find_elements(By.XPATH,"//td[@class='a-text-left mojo-field-type-title']")
for i in title:
     title_list.append(i.text)
     
title_list
```

```python
gross_list = []
gross =  fathi.find_elements(By.XPATH,"//td[@class='a-text-right mojo-field-type-money']")
for i in gross:
     gross_list.append(i.text)

gross_list
```

```python
rank_list = []
rank =  fathi.find_elements('xpath',"//td[@class='a-text-right mojo-header-column mojo-truncate mojo-field-type-rank']")
for i in rank:
     rank_list.append(i.text)

rank_list
```

```python
year_list = []
year =  fathi.find_elements('xpath',"//td[@class='a-text-left mojo-field-type-year']")
for i in year:
     year_list.append(i.text)

year_list
```

```python
movies= {'Rank': rank_list, 'Title': title_list, 'Lifetime Gross': gross_list,'Year':year_list } 
movie_data= pd.DataFrame(movies)
movie_data
```

```python
# save the dataframe as csv file
movie_data.to_csv('top_200_movies.csv',index=False)
print("Data saved to top_200_movies.csv")
```
